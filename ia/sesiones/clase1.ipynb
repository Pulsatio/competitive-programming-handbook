## ¬øQu√© es la programaci√≥n?
Programar es escribir instrucciones claras para que la computadora automatice tareas y convierta ideas en soluciones. Es similar a planificar una receta: necesitamos ingredientes (datos), pasos ordenados (algoritmos) y verificaci√≥n del resultado.

Se basa en 4 ideas clave que iremos desbloqueando en el curso:
1. **Creaci√≥n de datos:** guardar informaci√≥n como textos, n√∫meros o listas.
2. **Condicionales:** tomar decisiones dependiendo de una condici√≥n (`if`, `elif`, `else`).
3. **Bucles:** repetir acciones hasta completar una tarea (`for`, `while`).
4. **Funciones:** agrupar pasos en bloques reutilizables (`def mi_funcion():`).

## ¬øQu√© es la Inteligencia Artificial?
La IA combina programaci√≥n, estad√≠stica y matem√°ticas para que las m√°quinas aprendan patrones y tomen decisiones. Trabaja con conceptos como √°lgebra lineal (vectores, matrices), c√°lculo (cambios y optimizaci√≥n) y probabilidad/estad√≠stica (incertidumbre, inferencia). Gracias a esta base matem√°tica, podemos representar datos, ajustar modelos y evaluar qu√© tan bien funciona una predicci√≥n.

### Ramas principales y temas que estudian
1. **Machine Learning (aprendizaje autom√°tico):** regresi√≥n, clasificaci√≥n, √°rboles de decisi√≥n, `k`-means, validaci√≥n cruzada.
2. **Deep Learning (aprendizaje profundo):** redes neuronales, convoluciones, embeddings, retropropagaci√≥n, optimizadores como Adam.
3. **Procesamiento del Lenguaje Natural (PLN):** tokenizaci√≥n, modelos de lenguaje, an√°lisis de sentimiento, traducci√≥n autom√°tica.
4. **Visi√≥n por Computadora:** detecci√≥n de bordes, convoluciones 2D, clasificaci√≥n de im√°genes, segmentaci√≥n.
5. **Rob√≥tica:** cinem√°tica, controladores PID, planificaci√≥n de trayectorias, integraci√≥n de sensores.
6. **Sistemas Expertos:** reglas l√≥gicas, bases de conocimiento, motores de inferencia.
7. **L√≥gica Difusa:** conjuntos difusos, grados de pertenencia, reglas de inferencia suaves.
8. **Inteligencia Artificial Generativa (IAG):** GANs, modelos de difusi√≥n, transformers generativos para texto, imagen y audio.
